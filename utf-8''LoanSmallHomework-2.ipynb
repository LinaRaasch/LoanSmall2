```python
import pandas as pd
import numpy as np
import sklearn
location = 'loansmall.csv'
df = pd.read_csv(location)
df.head(None)
```

df.isnull().sum()

```python
medf = df.loc[df['addr_state'] == 'ME']
```

```python
medf.head(None)
```

```python
df1=medf.dropna(axis=1,how='all')
df1.head()
```

```python
grade = pd.pivot_table(medf, values=[ 'loan_amnt', 'int_rate' ], index=['addr_state', 'grade'],
                    aggfunc={'loan_amnt': [len, min, max, np.mean],
                            'int_rate': [min, max, np.mean]})
grade.head(None)
```

```python
df1.corr()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula = 'int_rate ~ annual_inc + mths_since_last_delinq + revol_util + loan_amnt + total_acc', data=df1).fit()
result.summary()
```
